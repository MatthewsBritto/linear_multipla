```python
import gradio as gr
import joblib
import pandas as pd
```

```python
modelo = joblib.load('./colesterol_model.pkl')
```

```python
def predict(grupo_sanguineo, fumante, nivel_atividade_fisica, idade, peso, altura):
    predicao_individual = {
    'grupo_sanguineo': grupo_sanguineo,
    'fumante': fumante,
    'nivel_atividade_fisica': nivel_atividade_fisica,
    'idade': idade,
    'peso': peso,
    'altura': altura,
    }
    predict_df = pd.DataFrame(predicao_individual, index=[0])
    colesterol = modelo.predict(predict_df)
    
    return float(colesterol[0])
```

```python
demo = gr.Interface(
  fn=predict,
  inputs=[
    gr.Radio(label='Grupo sanguineo', choices=['A', 'B', 'AB', 'O']),
    gr.Checkbox(label='Fumante'),
    gr.Radio(label='Nivel de atividade fisica', choices=['Baixo', 'Moderado', 'Alto']),
    gr.Slider(20,80,step=1, label='Idade'),
    gr.Slider(40, 160, step=1, label='Peso'),
    gr.Slider(150, 200, step=1, label='Altura'),
  ],
  outputs=gr.Number()
)
```

```python
demo.launch(debug=True)
```
